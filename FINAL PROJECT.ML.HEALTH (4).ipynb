# HOSPITAL_READMISSIONS

```python
import pandas as pd
```

Hospital_Readmissions dataset contains 25000 rows and 17 columns.It's aim is to build a machine learning model that predicts whether a patient will be readmitted to the hospital and when.That means the project focuses on analyzing hospital readmission data to understand patterns and factors contributing to patients being readmitted after discharge.

```python
a=pd.read_csv('hospital_readmissions.csv')
a
```

# LABEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
```

```python
j=LabelEncoder()
k=j.fit_transform(a['age'])
l=j.fit_transform(a['medical_specialty'])
m=j.fit_transform(a['diag_1'])
n=j.fit_transform(a['diag_2'])
o=j.fit_transform(a['diag_3'])
p=j.fit_transform(a['glucose_test'])
q=j.fit_transform(a['A1Ctest'])
r=j.fit_transform(a['change'])
s=j.fit_transform(a['diabetes_med'])
t=j.fit_transform(a['readmitted'])
a['age']=k
a['medical_specialty']=l
a['diag_1']=m
a['diag_2']=n
a['diag_3']=o
a['glucose_test']=p
a['A1Ctest']=q
a['change']=r
a['diabetes_med']=s
a['readmitted']=t
a
```

# CORRELATION

```python
a.corr()
```

```python
data=a.drop('n_procedures',axis='columns')
data
```

# TRAIN_TEST_SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.drop('readmitted',axis='columns')
x
```

```python
y=data.readmitted
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=40)
xtrain,xtest,ytrain,ytest
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
model1.predict(xtest)
```

```python
model1.score(xtest,ytest)
```

# CONFUSION MATRIX OF LOGISTIC REGRESSION

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
```

```python
y_pred=model1.predict(xtest)
y_pred
```

```python
cm=confusion_matrix(y_pred,ytest)
cm
```

```python
import seaborn as sn
```

```python
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
plt.figure(figsize=(20,7))
```

This confusion matrix shows the performance of linear model.It correctly classified 2101 true negatives, 1383 false positives, 571 false negatives and 945 true positives.

# VISUALIZATION

```python
import seaborn as sns
```

```python
sns.countplot(x='readmitted', data=data)
plt.title('Readmitted Class Distribution')
plt.xlabel('Readmitted (0=No, 1=Yes)')
plt.ylabel('count')
plt.show()
```

The bar chart shows that most patients were not readmitted (class 0), while a slightly smaller number were readmitted (class 1).The class distribution is nearly balanced.

```python
numeric_features = ['time_in_hospital', 'n_lab_procedures', 'n_medications', 'n_outpatient', 'n_inpatient', 'n_emergency']
data[numeric_features].hist(figsize=(10,8), bins=15)
plt.suptitle('Distribution of Numeric Features')
plt.show()
```

Most patients stayed in the hospital for a few days and had some lab tests and medicines.Very few patients had visited the hospital before, either as outpatients, inpatients, or in emergencies.Most of those values are close to zero.
